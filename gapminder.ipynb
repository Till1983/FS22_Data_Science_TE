## Step 2
#### Loading the tables

```python
df_fertility = pd.read_csv("children_per_woman_total_fertility.csv")
df_life_exp = pd.read_csv("life_expectancy_years.csv")
df_pop = pd.read_csv("population_total.csv")
```

```python
df_fertility = df_fertility.set_index("country")
df_fertility
```

```python
df_life_exp = df_life_exp.set_index("country")
df_life_exp
```

```python
df_pop = df_pop.set_index("country")
df_pop
```

## Step 3
#### Fertility and life expectancy for the year 2015

```python
fert_2015 = df_fertility["2015"]
fert_2015
```

```python
lifexp_2015 = df_life_exp["2015"]
lifexp_2015
```

```python
fert_lifexp_2015 = pd.concat([fert_2015, lifexp_2015], axis=1, ignore_index=True)
fert_lifexp_2015.head()
```

```python
fert_lifexp_2015 = fert_lifexp_2015.rename(columns={fert_lifexp_2015.columns[0]: 'fertility', fert_lifexp_2015.columns[1]: 'life_exp'})
```

```python
fert_lifexp_2015
```

## Step 4
#### Removing rows with missing values

```python
fert_lifexp_2015 = fert_lifexp_2015.dropna()
fert_lifexp_2015
```

## Step 5
#### Plotting fertility over life expectancy in 2015

```python
fertlife2015_plot = sns.scatterplot(x="fertility", y="life_exp", data=fert_lifexp_2015)
fertlife2015_plot
```
