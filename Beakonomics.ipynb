# Beakonomics

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
```

```python
df.head()
```

### 1. What does a typical penguin look like?

```python
sns.boxplot(data=df["bill_length_mm"], orient="h")
plt.title("Beak length (mm)")
```

```python
sns.boxplot(data=df["bill_depth_mm"], orient="h")
plt.title("Bill depth (mm)")
```

```python
sns.boxplot(data=df["flipper_length_mm"], orient="h")
plt.title("Flipper length (mm)")
```

```python
sns.boxplot(data=df["body_mass_g"], orient="h")
plt.title("Body mass (gram)")
```

### 2. What is the distribution of beak lengths and widths for different species?

The distribution of beak lengths for the species **Adelie** most closely resembles a normal distribution.

```python
adelie = df[df["species"] == "Adelie"]
sns.histplot(adelie, x="bill_length_mm")
plt.title("Distribution of beak lengths (Adelie)")
plt.xlabel("beak length (mm)")
```

```python
chinstrap = df[df["species"] == "Chinstrap"]
sns.histplot(chinstrap, x="bill_length_mm")
plt.title("Distribution of beak lengths (Chinstrap)")
plt.xlabel("beak length (mm)")
```

```python
gentoo = df[df["species"] == "Gentoo"]
sns.histplot(gentoo, x="bill_length_mm")
plt.title("Distribution of beak lengths (Gentoo)")
plt.xlabel("beak length (mm)")
```

### 3. Which species have long and slim beaks, which are rather short and wide?

**Adelies** have the shortest beaks, while those of **Chinstraps** and **Gentoos** are similarly long. **Gentoos** have notably slimmer beaks than the other two species.

```python
sns.scatterplot(df, x="bill_length_mm", y="bill_depth_mm", hue="species")
plt.xlabel("Bill length (mm)")
plt.ylabel("Bill depth (mm)")
```

### 4. Are the beak sizes in the most prevalent species correlated with any other feature of a penguin?

There is a positive correlation between beak length and body mass, as well as between beak length and flipper length for **Adelies** and **Gentoos**. The correlation between those variables in **Chinstraps** is still noticable, but weaker.

```python
sns.scatterplot(df, x="bill_length_mm", y="body_mass_g", hue="species")
plt.xlabel("Beak length (mm)")
plt.ylabel("Body mass (gram)")
```

```python
sns.scatterplot(df, x="bill_length_mm", y="flipper_length_mm", hue="species")
plt.xlabel("Beak length (mm)")
plt.ylabel("Flipper length (mm)")
```
